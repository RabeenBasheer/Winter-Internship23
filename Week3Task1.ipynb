```python
import pandas as pd
import numpy as np
salary_df = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
salary_df
```

```python
copy_df = salary_df.copy()
copy_df.describe()
```

```python
print(copy_df.shape)
print(copy_df.dtypes)
```

```python
copy_df.isnull()
```

```python
copy_df.isnull().sum()
```

```python
copy_df.Age.fillna(copy_df.Age.mean(), inplace = True)
copy_df['Age'] = copy_df.Age.round(decimals = 1)
copy_df
```

```python
copy_df.Salary
```

```python
copy_df['Salary'] = copy_df.Salary.str.replace("$", " ")
copy_df['Salary'] = copy_df.Salary
copy_df
```

```python
copy_df['Salary'] = copy_df.Salary.str.replace("k", "000")
copy_df['Salary'] = copy_df.Salary
copy_df
```

```python
copy_df['Location'] = copy_df.Location.str.replace(",In", " ")
copy_df['Location'] = copy_df.Location
copy_df
```

```python
copy_df['Location'] = copy_df.Location.str.replace(",Ny", " ")
copy_df['Location'] = copy_df.Location
copy_df
```

```python
copy_df['Location'] = copy_df.Location.str.replace(",Aus", " ")
copy_df['Location'] = copy_df.Location
copy_df
```

```python
copy_df['Location'] = copy_df.Location.str.replace(" In", " ")
copy_df['Location'] = copy_df.Location
copy_df
```

```python
copy_df['Location'] = copy_df.Location.str.replace(" Aus", " ")
copy_df['Location'] = copy_df.Location
copy_df
```

```python
copy_df.Rating.fillna(copy_df.Rating.mean(), inplace = True)
copy_df['Rating'] = copy_df.Rating.round(decimals = 1)
copy_df
```

```python
copy_df['Rating'] = copy_df['Rating'].abs()
copy_df
```

```python
copy_df['Rating'] = copy_df.Rating.replace("1.0", "0.0")
copy_df['Rating'] = copy_df.Rating
copy_df
```

```python
copy_df.rename(columns={"Easy Apply": "Apply"}, inplace=True)
copy_df
```

```python
copy_df['Apply'] = copy_df.Apply.replace("-1", "FALSE")
copy_df['Apply'] = copy_df.Apply
copy_df
```

```python
copy_df['Established'] = copy_df['Established'].replace(-1, np.nan)
copy_df
```

```python
copy_df['Established'].mode
```

```python
copy_df['Established'] = copy_df['Established'].replace(np.nan, copy_df['Established'].mode().iloc[0])
copy_df
```

```python
copy_df['Established'] = copy_df['Established'].astype(int)
copy_df
```
